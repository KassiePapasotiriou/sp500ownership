```
import ntpath

start_date = pd.to_datetime('2018-08-1')
end_date   = pd.to_datetime('2020-08-1')

def fix_df(d, name):
  d['Date'] =  pd.to_datetime(d['timestamp']).dt.date
  d.drop(['timestamp'], axis=1)
  d = d.groupby('Date').mean()
  d['ticker'] =  name
  d = d.loc[start_date:end_date]
  return d

d = pd.read_csv(rt_csv[0])
name = ntpath.basename(rt_csv[0])[:-4]
final_df = fix_df(d, name)

count = 0
for f in rt_csv[1:]:
  d = pd.read_csv(f)
  name = ntpath.basename(f)[:-4]
  final_df = final_df.append(fix_df(d, name))

  count += 1
  if count % 100 == 0:
    x = 0
    print(f"{count} stocks out of {len(rt_csv)} completed")
```
